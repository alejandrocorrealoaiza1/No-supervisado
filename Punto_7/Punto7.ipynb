```python
from keras.datasets import mnist
import numpy as np
import seaborn as sns
import matplotlib.patches as mpatches
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# CARGAR EL DATASET
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
#FILTRAR DATASET ENTRE CEROS Y OCHOS
X_train = X_train[(y_train==0)|(y_train==8)]
y_train_ = y_train[(y_train==0)|(y_train==8)]
X_test = X_test[(y_test==0)|(y_test==8)]
y_test = y_test[(y_test==0)|(y_test==8)]
```

```python
#REVISAR FILTRADO
plt.gray() 
fig, axes = plt.subplots(1,10,figsize=(20,200))
for k,ax in enumerate(axes):
    ax.imshow(X_train[k])
```

```python
#EL DATASET ES DE TRES DIMENSIONES, SE PASA A 2D
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1]*X_test.shape[2])
```

```python
#SE ESCALAN LOS DATOS
x_test = StandardScaler().fit_transform(X_test)
```

```python
#MODELO PCA
pca = PCA(n_components=2)
X_test_PCA = pca.fit_transform(X_test)
```

```python
#GRAFICAR PCA
fig,ax=plt.subplots(1,1,figsize=(9,9))
sns.scatterplot(x = X_test_PCA[:,0]
                , y = X_test_PCA[:,1]
                , hue = y_test
                , palette = 'Paired'
                , legend = 'full')
```

```python
# MODELO TSNE
tsne = TSNE(n_components=2)
X_test_TSNE = tsne.fit_transform(X_test)
```

```python
#GRAFICAR TSNE
fig,ax=plt.subplots(1,1,figsize=(9,9))
sns.scatterplot(x = X_test_TSNE[:,0]
                , y = X_test_TSNE[:,1]
                , hue = y_test
                , palette = 'Paired'
                , legend = 'full')
```
